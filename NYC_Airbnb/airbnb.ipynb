```python
import pandas as pd
```

```python
last_review = pd.read_csv("airbnb_last_review.csv")
price = pd.read_csv('airbnb_price.csv')
roomtype = pd.read_excel('airbnb_room_type.xlsx')

last_review_df = pd.DataFrame(last_review)
price_df = pd.DataFrame(price)
roomtype_df =pd.DataFrame(roomtype)
```

```python
last_review_df.count()
price_df.count()
roomtype_df.count()
```

```python
last_review_df.dropna()
last_review_df.count()
last_review_df.head(5)
```

```python
price_df.head(5)
```

```python
roomtype_df.head(5)
```

```python
rp = roomtype_df.set_index('listing_id').join(price_df.set_index('listing_id')).join(last_review_df.set_index(' listing_id'))
rp.count()
```

```python
rp[rp.isna().any(axis=1)]

```

```python
rp.dropna(inplace= True)
rp[rp.duplicated()]
rp.drop_duplicates(inplace=True)
rp.count()
```

```python
rp["last_review"] = pd.to_datetime(rp["last_review"])
rp.head()
```

```python
rp['price'] = [i.strip("dollars") for i in rp['price']]
rp['price'] = [int(i) for i in rp['price']]
rp['price'] = rp['price'].map("${:,.0f}".format)
```

```python
rp.isnull().sum()
```

```python
rp.to_csv("converted_airbnb_file.csv")
```
